````python
####################################################
# Import Numpy and Pandas library
# Write your code here
# 1 pts
####################################################
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
````

````python
####################################################
# Create a function named read_data
# - Has an input of filename, i.e. fname
# - Read the data as a Pandas DataFrame
# - Drop duplicate on `url`, keep the last ones
# - Drop col `names` and set `url` as index
# - Print the data shape
# - Return the dataset
# Write your code here (4 pts)
####################################################
def read_data(fname):
    """
    Read and preprocess the dataset.
    
    Parameters:
    - fname (str): The filename of the dataset to be read.
    
    Returns:
    - pd.DataFrame: The processed dataset.
    """
    with open(fname, 'r') as f:
        # Read the data
        df = pd.read_csv(f, sep=',', header=0)
        print(f'Data shape raw               : {df.shape}')

        # Drop duplicate on `url`, keep the last ones
        df.drop_duplicates(subset='url', keep='last')
        df_count_dup = df.duplicated(subset='url').sum()
        print(f'Number of duplicate url      : {df_count_dup}')
        print(f'Data shape after dropping    : {df.shape}')

        # Drop col `names` and set `url` as index
        df = df.drop(columns='name').set_index('url')        
        print(f'Data shape final             : {df.shape}')

    return df
````

````python
####################################################
# Create a function named split_train_test
# - Has two arguments
#   - X, the input (pd.Dataframe)
#   - y, the output (pd.Dataframe)
#   - test_size, the test size between 0-1 (float)
#   - seed, the random state (int)
# - Print the data shape after splitting
# - Return X_train, X_test, y_train, y_test
# - You can use an sklearn library to help you
# Write your code here
####################################################
def split_train_test(X, y, test_size, seed):
    """
    Split the input data (X) and target data (y) into train and test sets.

    Parameters:
    - X (pd.DataFrame): The input features.
    - y (pd.DataFrame or pd.Series): The target labels.
    - test_size (float): The proportion of the dataset to include in the test split (0-1).
    - seed (int): The random seed for reproducibility.

    Returns:
    - X_train (pd.DataFrame): Training input data.
    - X_test (pd.DataFrame): Testing input data.
    - y_train (pd.DataFrame or pd.Series): Training output data.
    - y_test (pd.DataFrame or pd.Series): Testing output data.
    """
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
        
    print(f'X train shape: {X_train.shape}')
    print(f'y train shape: {y_train.shape}')
    print(f'X test shape : {X_test.shape}')
    print(f'y test shape : {y_test.shape}\n')

    return X_train, X_test, y_train, y_test
````

- Create a function to fit a numerical features imputer

````python
####################################################
# Create function to fit & transform numerical imputers
# The fit function is called by num_imputer_fit
# - it needs 1 input, the data (pd.DataFrame)
# - the missing value is np.nan
# - the imputation strategy is median
# - it return the imputer
#
# The transform function is called by num_imputer_transform
# - it needs 2 input, data (pd.DataFrame) and imputer (sklearn object)
# - it return the imputed data in pd.DataFrame format
#
# Write your code here
####################################################
def num_imputer_fit(data):
    """
    Fit a numerical imputer.
    
    Parameters:
    - data (pd.DataFrame): The input DataFrame containing numerical data.
        
    Returns:
    - SimpleImputer: The fitted imputer.
    """
    
    fit_data = SimpleImputer(strategy='median').fit(data)

    return fit_data

def num_imputer_transform(data, imputer):
    """
    Transform the numerical data using a fitted imputer.
    
    Parameters:
    - data (pd.DataFrame): The input DataFrame containing numerical data.
    - imputer (SimpleImputer): The fitted imputer.
        
    Returns:
    - pd.DataFrame: The DataFrame with imputed values.
    """
    transform_data = pd.DataFrame(imputer.transform(data),
                                  columns=data.columns, index=data.index)

    return transform_data
````

````python
####################################################
# Create two functions to perform scaling & transform scaling
# The scaling is Standardization
# The first function is to fit the scaler, called by fit_scaler
# - You need an input, a data (pd.Dataframe)
# - You create a standardization scaler (please use sklearn)
# - Your output is the scaler
#
# The second function is to transform data using scaler, called by transform_scaler
# - There are two inputs, a data (pd.Dataframe), a scaler (sklearn object)
# - You scaled the data, then return the scaled data
# Write your code here
####################################################
def fit_scaler(data):
    """
    Fit a StandardScaler to the given data.
    
    Parameters:
    - data (pd.DataFrame): The input data to fit the scaler.
        
    Returns:
    - StandardScaler: The fitted scaler.
    """
    fit_data = StandardScaler().fit(data)

    return fit_data

def transform_scaler(data, scaled):
    """
    Transform the given data using the fitted scaler.
    
    Parameters:
    - data (pd.DataFrame): The input data to be transformed.
    - scaled (StandardScaler): The fitted scaler.
        
    Returns:
    - pd.DataFrame: Scaled data in the same format as the input DataFrame.
    """
    transformed_data = scaled.transform(data)
    transform_data = pd.DataFrame(transformed_data, columns=data.columns, index=data.index)

    return transform_data
````

#### 3.1. Preprare model evaluation function (10 pts)
---

- Before modeling, let's prepare two functions
  - `extract_cv_results`: to return the score and best param from hyperparameter search
  - `evaluate_model`: to return the RMSE of a model

````python
####################################################
# First, create a function to extract the CV results
# - The function name is extract_cv_results
# - It needs one input, called by `cv_obj` (a GridSearchCV sklearn object)
# - It returns three output:
#   1. the CV score on train set (float)
#   2. the CV score on valid set (float)
#   3. The best params (dictionary)
#
#
# Next, create a function to evaluate model called `rmse`
# - It needs 2 input
#   - y_actual, the actual output (pd.DataFrame or numpy array)
#   - y_pred, the predicted output (pd.DataFrame or numpy array)
# - You calculate the model performance using root mean squared error metrics
# - Then return the rmse
#
# Write your code here
####################################################
def extract_cv_results(cv_obj):
    """
    Extract cross-validation results from a GridSearchCV or RandomizedSearchCV object.
    
    Parameters:
    - cv_obj: GridSearchCV or RandomizedSearchCV object (fitted).

    Returns:
    - train_score: The best CV score on the training set (float).
    - valid_score: The best CV score on the validation set (float).
    - best_params: The best parameters found during the search (dictionary).
    """
    # Ensure the input is a fitted GridSearchCV/RandomizedSearchCV object
    if not hasattr(cv_obj, 'cv_results_'):
        raise ValueError("The input cv_obj must be a fitted GridSearchCV or RandomizedSearchCV object.")
    
    # Extract results
    train_scores = cv_obj.cv_results_['mean_train_score']
    valid_scores = cv_obj.cv_results_['mean_test_score']
    best_index = cv_obj.best_index_  # Index of the best parameter set
    
    # Extract best scores and parameters
    train_score = -train_scores[best_index]
    valid_score = -valid_scores[best_index]
    best_params = cv_obj.best_params_

    return train_score, valid_score, best_params

def rmse(y_actual, y_pred):
    """
    Calculate the Root Mean Squared Error (RMSE) between actual and predicted values.
    
    Parameters:
    - y_actual: Actual output values (pd.DataFrame or np.array).
    - y_pred: Predicted output values (pd.DataFrame or np.array).

    Returns:
    - rmse_value: The RMSE value (float).
    """
    if isinstance(y_actual, pd.DataFrame):
        y_actual = y_actual.values.flatten()
    if isinstance(y_pred, pd.DataFrame):
        y_pred = y_pred.values.flatten()
    
    rmse_value = np.sqrt(np.mean((y_actual - y_pred) ** 2))

    return rmse_value
````

````python
#########################################
# Write your code here
# Feel free to explore
#########################################
from sklearn.model_selection import RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score

def calculate_rmse(y_actual, y_pred):
    """Calculate Root Mean Squared Error"""
    return np.sqrt(mean_squared_error(y_actual, y_pred))

def refine_lasso_hyperparameters(X_train, X_valid, y_train, y_valid, initial_alpha=1000000.0):
    """
    Refine Lasso hyperparameters around the initial best alpha
    """
    # Create ranges of alpha values around the initial best alpha
    alpha_range1 = np.linspace(initial_alpha * 0.5, initial_alpha * 1.5, 100)
    alpha_range2 = np.logspace(np.log10(initial_alpha) - 1, np.log10(initial_alpha) + 1, 100)
    alphas = np.unique(np.concatenate([alpha_range1, alpha_range2]))
    
    results = []
    best_score = float('inf')
    best_model = None
    best_alpha = None
    
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('lasso', Lasso(max_iter=10000, random_state=42))
    ])
    
    print("Refining Lasso hyperparameters...")
    for alpha in alphas:
        # Set alpha
        pipeline.set_params(lasso__alpha=alpha)
        
        # Fit model
        pipeline.fit(X_train, y_train)
        
        # Calculate RMSE scores
        train_pred = pipeline.predict(X_train)
        valid_pred = pipeline.predict(X_valid)
        
        train_rmse = calculate_rmse(y_train, train_pred)
        valid_rmse = calculate_rmse(y_valid, valid_pred)
        
        # Store results
        results.append({
            'alpha': alpha,
            'train_rmse': train_rmse,
            'valid_rmse': valid_rmse,
            'n_features': np.sum(pipeline.named_steps['lasso'].coef_ != 0)
        })
        
        # Update best model if validation RMSE improves
        if valid_rmse < best_score:
            best_score = valid_rmse
            best_model = pipeline
            best_alpha = alpha
    
    results_df = pd.DataFrame(results)
    
    # Set up the figure with three horizontally stacked subplots
    fig, axes = plt.subplots(1, 3, figsize=(16, 6))

    # Plot RMSE vs Alpha (Training and Validation)
    axes[0].semilogx(results_df['alpha'], results_df['train_rmse'], 'b-', label='Training RMSE')
    axes[0].semilogx(results_df['alpha'], results_df['valid_rmse'], 'r-', label='Validation RMSE')
    axes[0].axvline(best_alpha, color='g', linestyle='--', label=f'Best Alpha: {best_alpha:.2f}')
    axes[0].set_xlabel('Alpha')
    axes[0].set_ylabel('Root Mean Squared Error')
    axes[0].set_title('Lasso Hyperparameter Tuning')
    axes[0].legend()
    axes[0].grid(True)

    # Plot Number of Selected Features vs Alpha
    axes[1].semilogx(results_df['alpha'], results_df['n_features'], 'm-', label='Number of Features')
    axes[1].axvline(best_alpha, color='r', linestyle='--', label=f'Best Alpha: {best_alpha:.2f}')
    axes[1].set_xlabel('Alpha')
    axes[1].set_ylabel('Number of Selected Features')
    axes[1].set_title('Number of Selected Features vs Alpha')
    axes[1].legend()
    axes[1].grid(True)

    # Get feature importance
    feature_importance = pd.DataFrame({
        'feature': X_train.columns,
        'coefficient': best_model.named_steps['lasso'].coef_,
        'importance': np.abs(best_model.named_steps['lasso'].coef_)
    }).sort_values('importance', ascending=False)

    # Plot Feature Importance
    sns.barplot(data=feature_importance.head(20), x='importance', y='feature', ax=axes[2])
    axes[2].set_title('Most Important Features')
    axes[2].set_xlabel('Importance')
    axes[2].set_ylabel('Feature')
    axes[2].tick_params(axis='x')

    # Adjust layout for better spacing between plots
    plt.tight_layout()

    # Show the combined figure
    plt.show()
    
    print("\nBest Model Results:")
    print(f"Best Alpha                  : {best_alpha:.6f}")
    print(f"Best Training Score (RMSE)  : {results_df.loc[results_df['alpha'] == best_alpha, 'train_rmse'].iloc[0] / 1e6:.4f} juta")
    print(f"Best Validation Score (RMSE): {results_df.loc[results_df['alpha'] == best_alpha, 'valid_rmse'].iloc[0] / 1e6:.4f} juta")
    print(f"Number of Selected Features : {int(results_df.loc[results_df['alpha'] == best_alpha, 'n_features'].iloc[0])}")
    
    return {
        'best_model': best_model,
        'best_alpha': best_alpha,
        'results': results_df,
        'feature_importance': feature_importance
    }

def evaluate_model(model, X, y, dataset_name=""):
    """
    Evaluate model performance using RMSE and R2
    """
    y_pred = model.predict(X)
    rmse = calculate_rmse(y, y_pred)
    r2 = r2_score(y, y_pred)
    
    print(f"\n{dataset_name} Performance:")
    print(f"RMSE     : {rmse / 1e6:.4f} juta")
    print(f"R2 Score : {r2:.6f}")
    
    return {'rmse': rmse, 'r2': r2}

# Refine hyperparameters
results = refine_lasso_hyperparameters(X_train_clean, X_valid_clean, y_train_dropped, y_valid, initial_alpha=1000000.0)

best_model = results['best_model']

# Evaluate on all datasets
train_metrics = evaluate_model(best_model, X_train_clean, y_train_dropped, "Training")
valid_metrics = evaluate_model(best_model, X_valid_clean, y_valid, "Validation")
test_metrics = evaluate_model(best_model, X_test_clean, y_test, "Test")
````
